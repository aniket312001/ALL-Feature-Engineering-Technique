```python
df = pd.read_csv('mercedes.csv',usecols=['X0','X1','X2','X3','X4','X5','X6','X8'])
df.head()
```

```python
for i in df.columns:
    print(f'No. of feature in {i} is {len(df[i].unique())}')
```

```python
df['X0'].value_counts().sort_values(ascending=False).head(10)
```

```python
df['X0'].value_counts().sort_values(ascending=False).head(10).index
```

```python
list_10 = list(df['X0'].value_counts().sort_values(ascending=False).head(10).index)

list_10
```

```python
for cat in list_10:
    df[cat] = np.where(df['X0']==cat,1,0)
```
